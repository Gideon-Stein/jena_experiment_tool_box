## Create a mapping from wrong to correct column name. 

- This product is used in the 1. transform script. 
- Can be updated if many faulty columns suddenly appear.
- Faulty columns and their corrections are detected in transform_raw.py and logged. 
- Check if many broken columns remain and update if necessary 

```python
import pandas as pd
import sys
sys.path.append("..")
from tools import get_all_paths
import pickle
```

## Original

```python
path = "header_file.csv"
head = pd.read_csv(path, encoding= 'unicode_escape')
tables, tables_list = get_all_paths()
```

```python
# This runs quite a bit...
additional = {}
for x in tables_list:
    try:
        data = pd.read_csv(x, encoding= 'unicode_escape')
        missing = []
        for y in data.columns: 
            if y not in head.columns: 
                missing.append(y)
        additional[x] = missing
    except: 
        pass
pickle.dump(additional, open("stats/additionalColumns_2003_2021.p", "wb"))
```

```python
col = pickle.load(open("stats/additionalColumns_2003_2021.p", "rb"))
```

```python
#Get all broken column names
names = []
for key in col: 
    for value in col[key]:
        names.append(value)
temp = []
for x in names: 
    if "Unnamed" in x: 
        pass
    else: 
        temp.append(x)
names = list(set(temp))
pickle.dump(names, open("stats/additionalColumnNames_2003_2021.p", "wb"))
```
