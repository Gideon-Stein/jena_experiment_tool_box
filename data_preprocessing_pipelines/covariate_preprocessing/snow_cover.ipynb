### Process snow data for Jena experiment site, extracted from DWD records

```python
# REPLACE to your data origin
d_path = "/home/datasets4/stein/jena_experiment_data_raw/"
out_path = "/home/datasets4/stein/jena_experiment_data_various_products/covariates_processed/"
```

```python
snow = pd.read_csv(d_path + "snow_dwd/tageswerte_RR_02444_18270101_20221231_hist/produkt_nieder_tag_18270101_20221231_02444.txt", sep=";",parse_dates=[1])
snow = snow[snow["MESS_DATUM"].dt.year > 2002]
snow.loc[snow["SH_TAG"] == -999,"SH_TAG"] = 0
snow.reset_index(inplace=True)
```

```python
start = 0
stack = []
for x in range(len(snow)):
    if snow["SH_TAG"].values[x] > 0: 
        if start == None: 
            start = x
    else: 
        if start != None:
            end = x-1
            stack.append([snow["MESS_DATUM"].values[start],snow["MESS_DATUM"].values[end]])
            start = None
```

```python
snow_dates = pd.DataFrame(stack, columns=["start", "end"])
```

```python
single_snow_days= pd.DataFrame(snow_dates[snow_dates ["start"] == snow_dates ["end"]]["start"].reset_index(drop=True))
snow_intervals = snow_dates.drop(index=snow_dates[snow_dates ["start"] == snow_dates["end"]].index).reset_index(drop=True)
```

```python
single_snow_days.to_csv(out_path + "single_snow_days.csv")
```

```python
snow_intervals.to_csv(out_path + "snow_intervals.csv")
```

```python
snow[["MESS_DATUM", "SH_TAG"]].to_csv(out_path + "snow_depth.csv")
```
